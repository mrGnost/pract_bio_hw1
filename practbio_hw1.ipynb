```python
# пункт 7

for seq_record in list(SeqIO.parse("bacteria.fasta", "fasta"))[:1]:
    print("Длина генома:", len(seq_record))
    nA = seq_record.seq.count("A")
    nG = seq_record.seq.count("G")
    nC = seq_record.seq.count("C")
    nT = seq_record.seq.count("T")
    print("Кол-во A:", nA)
    print("Кол-во G:", nG)
    print("Кол-во C:", nC)
    print("Кол-во T:", nT)
    pA = nA / len(seq_record)
    pG = nG / len(seq_record)
    pC = nC / len(seq_record)
    pT = nT / len(seq_record)
```

```python
pAAAA = pA**4
pAATT = pATAT = pATTA = pA**2 * pT**2
pAAAA, pAATT, pATAT, pATTA
```

```python
import sys, time

genome_len = 1576758
freqs = []
for i in range(3, 500):
  os.system('jellyfish count -m {} -s 100M -t 10 -C bacteria.fasta -o {}mers.txt'.format(i, i))
  singles = int(os.popen("jellyfish dump {}mers.txt | grep -x '>1' | wc -l".format(i)).read())
  all = int(os.popen("jellyfish dump {}mers.txt | grep --count '>'".format(i)).read())
  freqs.append(singles / all * 100)
  print(i, freqs[-1])
  if freqs[-1] > 99:
    break

# Получается первый к с 99% уникальными кмерами равен 272
```

```python
import matplotlib.pyplot as plt

plt.plot(range(3, 273), freqs)

# Есть мысль, что в геноме может быть дублированный ген,
# из-за чего настолько длинные кмеры в нем не уникальны.
```
